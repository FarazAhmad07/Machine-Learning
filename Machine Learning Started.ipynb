```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, train_test_split
from sklearn import metrics

# Load data from CSV file
file_path = 'F:/DataUpdated.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Separate features (X) and target variable (y)
X = data.drop('RV', axis=1)  # Features
y = data['RV']  # Target variable

# Define KFold cross-validator
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Create a Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

fold = 1
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Training the model
    clf.fit(X_train, y_train)
    
    # Training precision for fold
    y_train_pred = clf.predict(X_train)
    train_precision = metrics.precision_score(y_train, y_train_pred, average='weighted')
    print(f"Training Precision for fold {fold}: {train_precision:.4f}")
    
    # Testing precision for fold
    y_pred = clf.predict(X_test)
    test_precision = metrics.precision_score(y_test, y_pred, average='weighted')
    print(f"Testing Precision for fold {fold}: {test_precision:.4f}")
    
    # Testing recall for fold
    test_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    print(f"Testing Recall for fold {fold}: {test_recall:.4f}\n")
    
    fold += 1
```

Explanation:

param_grid: This variable holds a dictionary that defines a grid of hyperparameters to search through during grid search.
C: The C parameter in SVM controls the trade-off between a smooth decision boundary and classifying training points correctly. In this grid, values [0.1, 1, 10, 100] for C are considered.
gamma: The gamma parameter defines the influence of a single training example. Higher values of gamma lead to more complex decision boundaries. In this grid, values [1, 0.1, 0.01, 0.001] for gamma are considered.
kernel: The kernel parameter in SVM specifies the type of hyperplane used to separate the data. Different kernel functions can be used ('rbf' for radial basis function, 'linear' for a linear kernel, 'poly' for polynomial, 'sigmoid' for sigmoid kernel).
The grid search technique then explores different combinations of these hyperparameters (C, gamma, kernel) and evaluates each combination using cross-validation to determine the best-performing set of hyperparameters based on a specified scoring metric (e.g., accuracy, precision, recall). This process helps in finding the hyperparameter values that yield the best performance for the given dataset and model.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_predict, cross_val_score, StratifiedKFold
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a Naive Bayes Classifier (Gaussian Naive Bayes)
clf = GaussianNB()

# Perform cross-validation
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Initialize lists to store evaluation metrics
train_accuracy_list = []
train_precision_list = []
train_recall_list = []
train_f1_score_list = []
test_accuracy_list = []
test_precision_list = []
test_recall_list = []
test_f1_score_list = []

# Calculate metrics for each fold in cross-validation
for train_index, test_index in cv.split(X_scaled, y):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Train the classifier
    clf.fit(X_train, y_train)
    
    # Make predictions on training set
    y_pred_train = clf.predict(X_train)
    # Calculate metrics for training set
    train_accuracy = metrics.accuracy_score(y_train, y_pred_train)
    train_precision = metrics.precision_score(y_train, y_pred_train, average='weighted')
    train_recall = metrics.recall_score(y_train, y_pred_train, average='weighted')
    train_f1_score = metrics.f1_score(y_train, y_pred_train, average='weighted')

    # Make predictions on test set
    y_pred_test = clf.predict(X_test)
    # Calculate metrics for test set
    test_accuracy = metrics.accuracy_score(y_test, y_pred_test)
    test_precision = metrics.precision_score(y_test, y_pred_test, average='weighted')
    test_recall = metrics.recall_score(y_test, y_pred_test, average='weighted')
    test_f1_score = metrics.f1_score(y_test, y_pred_test, average='weighted')

    # Append metrics to respective lists
    train_accuracy_list.append(train_accuracy)
    train_precision_list.append(train_precision)
    train_recall_list.append(train_recall)
    train_f1_score_list.append(train_f1_score)
    test_accuracy_list.append(test_accuracy)
    test_precision_list.append(test_precision)
    test_recall_list.append(test_recall)
    test_f1_score_list.append(test_f1_score)

# Calculate mean and standard deviation of metrics
mean_train_accuracy = np.mean(train_accuracy_list)
mean_train_precision = np.mean(train_precision_list)
mean_train_recall = np.mean(train_recall_list)
mean_train_f1_score = np.mean(train_f1_score_list)
mean_test_accuracy = np.mean(test_accuracy_list)
mean_test_precision = np.mean(test_precision_list)
mean_test_recall = np.mean(test_recall_list)
mean_test_f1_score = np.mean(test_f1_score_list)

std_train_accuracy = np.std(train_accuracy_list)
std_train_precision = np.std(train_precision_list)
std_train_recall = np.std(train_recall_list)
std_train_f1_score = np.std(train_f1_score_list)
std_test_accuracy = np.std(test_accuracy_list)
std_test_precision = np.std(test_precision_list)
std_test_recall = np.std(test_recall_list)
std_test_f1_score = np.std(test_f1_score_list)

# Display results
print(f"Training Metrics:")
print(f"Mean Accuracy: {mean_train_accuracy:.4f} (+/- {std_train_accuracy * 2:.4f})")
print(f"Mean Precision: {mean_train_precision:.4f} (+/- {std_train_precision * 2:.4f})")
print(f"Mean Recall: {mean_train_recall:.4f} (+/- {std_train_recall * 2:.4f})")
print(f"Mean F1 Score: {mean_train_f1_score:.4f} (+/- {std_train_f1_score * 2:.4f})")

print(f"\nTesting Metrics:")
print(f"Mean Accuracy: {mean_test_accuracy:.4f} (+/- {std_test_accuracy * 2:.4f})")
print(f"Mean Precision: {mean_test_precision:.4f} (+/- {std_test_precision * 2:.4f})")
print(f"Mean Recall: {mean_test_recall:.4f} (+/- {std_test_recall * 2:.4f})")
print(f"Mean F1 Score: {mean_test_f1_score:.4f} (+/- {std_test_f1_score * 2:.4f})")
```
